```python
# Cleaning the missing data

# More than 30% null valued columns

nullvalues = df.isnull().sum()
nullvalues = nullvalues[nullvalues.values > (0.3 * len(nullvalues))]
len(nullvalues)
```

```python
# There are 55374 XNA columns, we will replace them with NaN

df.ORGANIZATION_TYPE = df.ORGANIZATION_TYPE.replace('XNA', np.NaN)
```
